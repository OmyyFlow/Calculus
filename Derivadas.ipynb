```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Derivada de una funci贸n por definicion:**
$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}  
$$

```python
def derivada(f,x,h):
  return (f(x+h) - f(x))/h
```

```python
n = 10
x = np.linspace(-n,n,n*10)
系 = 1e-10
```

```python
y = x**3
y_derivada = derivada(lambda x: x**3,x,系)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,y,label='y = x^3')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,3*x**2,label='derivada analitica y = 3x^2')
sns.scatterplot(x=x, y= y_derivada,color='red', label = 'derivada por definici贸n')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
y2 = np.sin(x)
y2_der = derivada(lambda x: np.sin(x),x,系)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,y2,label='y = sin(x)')
#plt.grid(True)
plt.hlines(y=0,xmin=-n,xmax=n,linestyles='-',color='gray')
plt.vlines(x=0,ymin=-1,ymax=1,linestyles='-',color='gray')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,np.cos(x),label='y=cos')
sns.scatterplot(x=x,y=y2_der,color='red', label = 'y_der')
#plt.grid(True)
plt.hlines(y=0,xmin=-n,xmax=n,linestyles='-',color='gray')
plt.vlines(x=0,ymin=-1,ymax=1,linestyles='-',color='gray')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
y3 = x
y3_der = derivada(lambda x: x,x,系)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,y3,label='y = x')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
xlim = plt.gca().get_xlim()
ylim = plt.gca().get_ylim()

plt.subplot(1,2,2)
sns.scatterplot(x=x,y=y3_der,color='red', label = 'y_der')
plt.plot(x, np.ones_like(x),label='y=1')
plt.xlim(xlim)
plt.ylim(ylim)

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-2,2,400, endpoint=False)
y4 = x*(x+1)*(x-1)*(x+2)*(x-2)
y4_der = derivada(lambda x: x*(x+1)*(x-1)*(x+2)*(x-2),x,系)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x,y4,label='y = x*(x+1)*(x-1)*(x+2)*(x-2)')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
#xlim = plt.gca().get_xlim()
#ylim = plt.gca().get_ylim()

plt.subplot(1,2,2)
sns.scatterplot(x=x,y=y4_der,color='red', label = 'y_der')
plt.plot(x, (5*x**4 - 15*x**2 +4),label='y=5x^4 - 15x^2 +4')
#plt.xlim(xlim)
#plt.ylim(ylim)

plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```

**Linearizacion**
Recta tangente de $f(x)$ en un punto $x_0$
$$y = f'(x_0)(x-x_0) + f(x_0)$$

```python
punto_x0 = -0.6
x0 = int(punto_x0*100 + 200)
recta_tan = y4_der[x0]*(x-x[x0]) + y4[x0]

plt.figure(figsize=(10,6))
plt.plot(x, y4, label = 'y = x*(x+1)*(x-1)*(x+2)*(x-2)')
plt.plot(x, recta_tan, label = 'Recta tangente en x_0 = 0', color = 'red')
plt.plot(punto_x0, y4[x0], 'bo', label = 'Punto (x_0,y_0)')
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-4,4)
plt.show()
```
